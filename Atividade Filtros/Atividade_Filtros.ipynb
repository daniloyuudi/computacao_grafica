# Filtro 1

```python
imagem = cv2.imread('entrada.png')
```

```python
for y in range(0, imagem.shape[0]):
  for x in range(0, imagem.shape[1]):
    (b, g, r) = imagem[y, x]
    imagem[y, x] = (b / 1.2, g / 2, r / 1.5)
```

```python
cv2_imshow(imagem)
```

# Filtro 2

```python
imagem = cv2.imread('entrada.png')
```

```python
for y in range(0, imagem.shape[0]):
  for x in range(0, imagem.shape[1]):
    (b, g, r) = imagem[y, x]
    imagem[y, x] = (min(255, b * 1.8), min(255, g * 1.8), min(255, r * 1.8))
```

```python
cv2_imshow(imagem)
```
