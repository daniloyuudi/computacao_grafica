```python
# Importação das bibliotecas
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
# Leitura da imagem com a função imread()
imagem = cv2.imread('entrada_2.jpg')
```

```python
# Variável contendo os codigos das cores
vermelho = (0, 0, 255)
verde = (0, 255, 0)
azul = (255, 0, 0)
```

```python
# Cria uma linha verde e vermelho
cv2.line(imagem, (0, 0), (100, 200), verde)
cv2.line(imagem, (300, 200), (150, 150), vermelho, 5)
```

```python
# Cria um retângulo azul e verde
cv2.rectangle(imagem, (20, 20), (120, 120), azul, 10)
cv2.rectangle(imagem, (200, 50), (225, 125), verde, -1)
```

```python
# Cria vários círculos vermelhos com variação do raio
(X, Y) = (imagem.shape[1] // 2, imagem.shape[0] // 2)
for raio in range(0, 175, 15):
  cv2.circle(imagem, (X, Y), raio, vermelho)
```

```python
cv2_imshow(imagem)
```
